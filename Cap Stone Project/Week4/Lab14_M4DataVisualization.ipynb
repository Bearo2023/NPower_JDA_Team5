## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""

df = pd.read_sql_query(QUERY,conn)

import matplotlib.pyplot as plt

# create a histogram
plt.hist(df['ConvertedComp'], bins=10)

# add labels and title
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')

# display the plot
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns
# create a box plot of the 'ConvertedComp' column using seaborn
sns.boxplot(y=df['Age'])

# display the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

# create a scatter plot
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)

# add labels and title
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')

# display the plot
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
sns.set(style='whitegrid')

# create the bubble plot
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(50, 200), alpha=0.7)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import matplotlib.pyplot as plt

# subset the data for age group 30-35
df_age = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# calculate median work week hours and code review hours for each age
df_median = df_age.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# create stacked bar chart
ax = df_median.plot(kind='bar', stacked=True, figsize=(10,6))

# set axis labels and title
ax.set_xlabel('Age')
ax.set_ylabel('Hours')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')

# display the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# group data by age and calculate median ConvertedComp for each age group
age_median = df[df['Age'].between(25, 30)].groupby('Age')['ConvertedComp'].median()

# create line chart using pandas plot function
ax = age_median.plot(kind='line', figsize=(10,6), grid=True, color='blue')

# set chart title and axis labels
ax.set_title('Median ConvertedComp by Age', fontsize=18)
ax.set_xlabel('Age', fontsize=14)
ax.set_ylabel('Median ConvertedComp', fontsize=14)

# show the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

import seaborn as sns

# Create a count plot with horizontal bars
plt.figure(figsize=(10, 6))
sns.countplot(y="MainBranch", data=df, order=df['MainBranch'].value_counts().index)
plt.xlabel("Count")
plt.ylabel("Main Branch")
plt.title("Number of respondents by Main Branch")
plt.show()
```
